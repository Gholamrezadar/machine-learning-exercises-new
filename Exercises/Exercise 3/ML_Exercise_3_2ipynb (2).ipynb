## Load data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import seaborn as sns; sns.set()

import sklearn
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import confusion_matrix, plot_confusion_matrix, accuracy_score, f1_score, ConfusionMatrixDisplay
from xgboost import XGBClassifier

sns.set_style("dark")
```

```python
# Loading the iris dataset
data_train = pd.read_csv("data_train.csv",names=range(17), index_col=False)
data_test = pd.read_csv("data_test.csv",names=range(17), index_col=False)

y_train = data_train.iloc[:,-1]
X_train = data_train.iloc[:,:-1]

y_test = data_test.iloc[:,-1]
X_test = data_test.iloc[:,:-1]
```

```python
X_train.head()
```

## Helper Functions

## a) Random Forest

### Training

```python
clf = RandomForestClassifier(max_depth=3,n_estimators=15,max_features=3, random_state=0)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

### Accuracy

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of the random forest model is {accuracy}")
f1 = f1_score(y_test, y_pred, average="weighted")
print(f"f1-score of the random forest model is {f1}")
```

### Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred, labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=clf.classes_)
disp.plot()
plt.show()
```

## b) Adaboost


```python
clf = AdaBoostClassifier(n_estimators=10, random_state=0, learning_rate=0.3)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

### Accuracy

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of the model with 10 classifiers is {accuracy}")
f1 = f1_score(y_test, y_pred, average="weighted")
print(f"f1-score of the model with 10 classifiers is {f1}")
```

### Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred, labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=clf.classes_)
disp.plot()
plt.show()
```
